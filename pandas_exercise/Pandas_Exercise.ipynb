```python
company=df[["company"]].drop_duplicates().reset_index(drop=True)
```

```python
company["reliability_index"]=np.random.randint(50,100,size=16)
company
```

```python
df.head()
```

```python
df.merge(company)
```

## inner merge(join)

```python
filtree=company["company"]!="alfa-romero"
company2=company[filtree]
```

```python
df.merge(company2) ##data setein birinde alfa-romero olmadığı için eşleşen satırları inner merge yaptı.
```

## left/right outer merge  

```python
df.merge(company2,how="left")
```

```python
df.merge(company2,how="right")
```

## outer merge(join)

```python
company3 = company2.append({'company':'tata','reliability_index':67}, ignore_index=True)
```

```python
df.merge(company3,how="outer").tail()
```
